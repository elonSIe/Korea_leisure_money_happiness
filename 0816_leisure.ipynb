```python
plt.style.use('fivethirtyeight')

plt.rcParams['figure.figsize'] = (10, 5)

sns.barplot(data = df, x = 'house_income', y = 'cost', hue = 'online', palette = 'Set3')
plt.title('leisure cost by house income', fontsize = 20, fontweight = 100)
label = plt.legend(loc = 'best')
label.get_texts()[0].set_text('Offline')
label.get_texts()[1].set_text('Online');
```

```python
plt.rcParams['figure.figsize'] = (10, 5)

sns.barplot(data = df, x = 'age', y = 'cost', palette = 'Set3')
plt.title('leisure cost by age group', fontsize = 20, fontweight = 100);
```

```python
plt.style.use('_classic_test')

colors = plt.cm.cool(np.linspace(0, 30, 200))
pie = df['age'].value_counts().plot.pie(colors = colors,
                                      figsize = (5, 5))

plt.title('Age Group',
          fontsize = 20, fontweight = 30)
plt.axis('off')
label = plt.legend(loc = 'center')
label.get_texts()[0].set_text('50-59Y')
label.get_texts()[1].set_text('40-49Y')
label.get_texts()[2].set_text('30-39Y')
label.get_texts()[3].set_text('20-29Y')
label.get_texts()[4].set_text('60-69Y')
label.get_texts()[5].set_text('70-79Y')
label.get_texts()[6].set_text('10-19Y');
```

```python
plt.style.use('_classic_test')

colors = plt.cm.cool(np.linspace(0, 50, 100))
df['sex'].value_counts().plot.pie(colors = colors,
                                      figsize = (5, 5))

plt.title('Sex',
          fontsize = 20, fontweight = 30)
plt.axis('off')
label = plt.legend(loc = 'center')
label.get_texts()[0].set_text('Male')
label.get_texts()[1].set_text('Female');
```
